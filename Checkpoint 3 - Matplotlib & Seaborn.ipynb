# **OUTLIERS**

Outliers adalah nilai yang jauh berbeda dari nilai lainnya dalam dataset. Nilai Outlier bisa jauh lebih rendah atau lebih tinggi. Outlier bisa terjadi karena berbagai alasan seperti faktor kesalahan maupun kejadian lain yang tidak terduga.

# **MELAKUKAN PENGECEKAN OUTLIERS**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

MENGATASI OUTLIERS DENGAN MALAKUKAN IMPUTASI

```python
columns_to_impute = ["CGPA100", "CGPA300", "CGPA400"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Melakukan pengecekan ulang

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Dari scatter plot yang ditampilkan, kita bisa melihat bagaimana hubungan antara IPK setiap tahun (CGPA100-CGPA400) dengan IPK akhir (CGPA). Beberapa pola awal yang terlihat:

1. **CGPA100 vs CGPA**

3. **CGPA300 vs CGPA**  
   Sebagian besar data terlihat lebih terkonsentrasi, yang berarti mahasiswa yang memiliki IPK tinggi di tahun ketiga cenderung mempertahankan nilai tersebut hingga lulus. Dengan kata lain, performa akademik di tahun ketiga bisa menjadi indikator yang cukup kuat untuk memprediksi IPK akhir.
   
4. **CGPA400 vs CGPA**  
   IPK di tahun keempat punya pengaruh besar terhadap IPK akhir. Mahasiswa yang punya IPK tinggi di tahun terakhir biasanya juga lulus dengan IPK yang tinggi. Ini masuk akal karena di tahun terakhir, nilai-nilai sudah hampir final, dan sedikit perubahan bisa berdampak langsung pada IPK keseluruhan.

Jadi, semakin mendekati tahun akhir, hubungan antara IPK per tahun dengan IPK akhir semakin kuat.

Heatmap korelasi menunjukkan hubungan antara IPK tiap tahun (CGPA100-CGPA400) dengan IPK akhir (CGPA):

- CGPA400 memiliki korelasi tertinggi dengan CGPA, yang berarti IPK tahun keempat sangat menentukan IPK akhir mahasiswa.
- CGPA100 memiliki korelasi paling rendah, menunjukkan bahwa IPK tahun pertama tidak selalu mencerminkan hasil akhir mahasiswa.
- CGPA200 dan CGPA300 memiliki korelasi yang cukup tinggi, artinya performa di tahun kedua dan ketiga juga berkontribusi signifikan terhadap IPK akhir.
